```python
#ポアソン分布の描画（棒グラフ）
import numpy as np
import matplotlib.pylab as plt
import math

def poisson(lambda_, k):
    k = int(k)
    result = (lambda_**k) * (math.exp(-lambda_))  / math.factorial(k)
    return result

x =  np.arange(1, 20, 1)
y1= [poisson(1,i) for i in x]
y2= [poisson(2,i) for i in x]
y3= [poisson(3,i) for i in x]
y4= [poisson(5,i) for i in x]
y5= [poisson(10,i) for i in x]

plt.bar(x, y1, align="center", width=0.4, color="black"
                ,alpha=1, label="Poisson λ= %d" % 1)

plt.bar(x, y2, align="center", width=0.4, color="red"
                ,alpha=0.5, label="Poisson λ= %d" % 2)

plt.bar(x, y3, align="center", width=0.4, color="green"
                ,alpha=0.5, label="Poisson λ= %d" % 3)

plt.bar(x, y4, align="center", width=0.4, color="blue"
                ,alpha=0.5, label="Poisson λ= %d" % 5)

plt.bar(x, y5, align="center", width=0.4, color="cyan"
                ,alpha=0.5, label="Poisson λ= %d" % 10)

plt.legend()
plt.show()
```

```python
#ポアソン分布の描画（折れ線グラフ）
import numpy as np
import matplotlib.pylab as plt
import math

def poisson(lambda_, k):
    k = int(k)
    result = (lambda_**k) * (math.exp(-lambda_))  / math.factorial(k)
    return result

x =  np.arange(0, 21, 1)
y1= [poisson(1,i) for i in x]
y2= [poisson(2,i) for i in x]
y3= [poisson(3,i) for i in x]
y4= [poisson(5,i) for i in x]
y5= [poisson(10,i) for i in x]

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.plot(x, y1, label='λ=1', marker='o')
ax.plot(x, y2, label='λ=2', marker='o')
ax.plot(x, y3, label='λ=3', marker='o')
ax.plot(x, y4, label='λ=5', marker='o')
ax.plot(x, y5, label='λ=10', marker='o')

ax.legend()
ax.set_xlabel("k")
ax.set_ylabel("P(X=k)")
ax.set_ylim(0, 0.4)

plt.show()
```

```python
#幾何分布の描画（折れ線グラフ）
import numpy as np
import matplotlib.pylab as plt
import math

def geo(p, k):
    p = float(p)
    k = int(k)
    result = p*((1-p)**(k-1))
    return result

x =  np.arange(1, 10, 1)
y1= [geo(0.05,i) for i in x]
y2= [geo(0.1,i) for i in x]
y3= [geo(0.2,i) for i in x]
y4= [geo(0.5,i) for i in x]
y5= [geo(0.9,i) for i in x]

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.plot(x, y1, label='p=0.05', marker='o')
ax.plot(x, y2, label='p=0.1', marker='o')
ax.plot(x, y3, label='p=0.2', marker='o')
ax.plot(x, y4, label='p=0.5', marker='o')
ax.plot(x, y5, label='p=0.9', marker='o')

ax.legend()
ax.set_xlabel("k")
ax.set_ylabel("P(X=k)")
ax.set_ylim(0, 1.0)

plt.show()
```

```python
#幾何分布の累積分布関数の描画（折れ線グラフ）
#直感的なグラフはこっちの方！！
import numpy as np
import matplotlib.pylab as plt
import math

def geo(p, k):
    p = float(p)
    k = int(k)
    result = p*((1-p)**(k-1))
    return result

def geo_sum(p, k):
    p = float(p)
    k = int(k)
    x =  np.arange(1, k, 1)
    y = [geo(p, i) for i in x]
    result = sum(y)
    return result

x =  np.arange(1, 15, 1)
y1= [geo_sum(0.05,i) for i in x]
y2= [geo_sum(0.1,i) for i in x]
y3= [geo_sum(0.2,i) for i in x]
y4= [geo_sum(0.5,i) for i in x]
y5= [geo_sum(0.9,i) for i in x]

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.plot(x, y1, label='p=0.05')
ax.plot(x, y2, label='p=0.1')
ax.plot(x, y3, label='p=0.2')
ax.plot(x, y4, label='p=0.5')
ax.plot(x, y5, label='p=0.9')

ax.legend()
ax.set_xlabel("k")
ax.set_ylabel("P(X=k)")
ax.set_ylim(0, 1.0)

plt.show()
```

```python
#正規分布の描画①
import matplotlib.pyplot as plt
import numpy as np
import math

a = input('平均の値は？：')
s = input('標準偏差の値は？：')
a = float(a)
s = float(s)
p = np.linspace(a-10, a+10, 200)   # linspace(min, max, N) で範囲 min から max を N 分割します
q = (np.exp((-(p-a)**2)/(2*(s**2))))/(math.sqrt(2*math.pi)*s)
plt.plot(p, q)
plt.show()
```

```python
#正規分布の描画②
import matplotlib.pyplot as plt
import numpy as np
import math

a = 60
p = np.linspace(a-10, a+10, 200)   # linspace(min, max, N) で範囲 min から max を N 分割します

def normal(s):
    return (np.exp((-(p-a)**2)/(2*(s**2)))) / (math.sqrt(2*math.pi)*s)

y1, = plt.plot(p, normal(1))
y2, = plt.plot(p, normal(2))
y3, = plt.plot(p, normal(5))
y4, = plt.plot(p, normal(10))
y5, = plt.plot(p, normal(20))
plt.legend([y1, y2, y3, y4, y5], ['σ=1', 'σ=2', 'σ=5', 'σ=10', 'σ=20'])
plt.show()
```

```python
#指数分布の描画
import matplotlib.pyplot as plt
import numpy as np
import math

p = np.linspace(0, 15, 150)   # linspace(min, max, N) で範囲 min から max を N 分割します

def index(s):
    return s * np.exp(-s*p)

y1, = plt.plot(p, index(0.1))
y2, = plt.plot(p, index(0.5))
y3, = plt.plot(p, index(1))
y4, = plt.plot(p, index(2))
y5, = plt.plot(p, index(5))
plt.legend([y1, y2, y3, y4, y5], ['λ=0.1', 'λ=0.5', 'λ=1', 'λ=2', 'λ=5'])
plt.ylim(0, 0.4)
plt.show()
```

```python
#指数分布の累積分布関数の描画
import matplotlib.pyplot as plt
import numpy as np
import math

p = np.linspace(0, 15, 150)   # linspace(min, max, N) で範囲 min から max を N 分割します

def index_c(s):
    return 1 - np.exp(-s*p)

y1, = plt.plot(p, index_c(0.1))
y2, = plt.plot(p, index_c(0.5))
y3, = plt.plot(p, index_c(1))
y4, = plt.plot(p, index_c(2))
y5, = plt.plot(p, index_c(5))
plt.legend([y1, y2, y3, y4, y5], ['λ=0.1', 'λ=0.5', 'λ=1', 'λ=2', 'λ=5'])
plt.ylim(0, 1)
plt.show()
```
