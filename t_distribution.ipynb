```python
#自由度を決めてt分布を出力する
import matplotlib.pyplot as plt
import numpy as np
import math

def normal(t):
    return (np.exp((-t**2)/2)) / (math.sqrt(2*math.pi))

def student(v, t):
    return ((math.gamma((v+1)/2))/(math.sqrt(v*math.pi)*math.gamma(v/2)))*((1+(t**2)/v)**(-(v+1)/2))

v = input('自由度を入力してください：')
v = int(v)

t = np.linspace(-5, 5, 200)   # linspace(min, max, N) で範囲 min から max を N 分割します
s = student(v, t)
y1, = plt.plot(t, s)
y5, = plt.plot(t, normal(t),linestyle = "dashed")
plt.legend([y1, y5], ['v='+str(v), 'normal'])
plt.show()
```

```python
#自由度を大きくするほどt分布は正規分布に収束する
import matplotlib.pyplot as plt
import numpy as np
import math

def normal(t):
    return (np.exp((-t**2)/2)) / (math.sqrt(2*math.pi))

def student(v, t):
    return ((math.gamma((v+1)/2))/(math.sqrt(v*math.pi)*math.gamma(v/2)))*((1+(t**2)/v)**(-(v+1)/2))

t = np.linspace(-5, 5, 200)   # linspace(min, max, N) で範囲 min から max を N 分割します

y1, = plt.plot(t, student(1, t))
y2, = plt.plot(t, student(2, t))
y3, = plt.plot(t, student(5, t))
y4, = plt.plot(t, student(10, t))
y5, = plt.plot(t, normal(t), linestyle = "dashed")
plt.legend([y1, y2, y3, y4, y5], ['v=1', 'v=2', 'v=5', 'v=10', 'normal'])
plt.show()
```
