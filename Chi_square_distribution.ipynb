```python
#自由度を決めてカイ二乗分布を出力する
import matplotlib.pyplot as plt
import numpy as np
import math

def chi(v, t):
    return (np.exp(-t/2)*(t**(v/2-1)))/((2**(v/2))*math.gamma(v/2))

v = input('自由度を入力してください：')
v = int(v)

t = np.linspace(0, 10, 200)   # linspace(min, max, N) で範囲 min から max を N 分割します
s = chi(v, t)
y1, = plt.plot(t, s)
plt.ylim(-0.1,1)
plt.show()
```

```python
#自由度を大きくするほどt分布は正規分布に収束する
import matplotlib.pyplot as plt
import numpy as np
import math

def chi(v, t):
    return (np.exp(-t/2)*(t**(v/2-1)))/((2**(v/2))*math.gamma(v/2))

t = np.linspace(0, 10, 200)   # linspace(min, max, N) で範囲 min から max を N 分割します

y1, = plt.plot(t, chi(1, t))
y2, = plt.plot(t, chi(2, t))
y3, = plt.plot(t, chi(3, t))
y4, = plt.plot(t, chi(5, t))
y5, = plt.plot(t, chi(10, t))
plt.legend([y1, y2, y3, y4, y5], ['k=1', 'k=2', 'k=3', 'k=5', 'k=10'])
plt.ylim(-0.1,1)
plt.show()
```
